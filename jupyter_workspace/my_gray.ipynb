```python
from PIL import Image #python图像处理库
import numpy
```

```python
img_path = r'D:\photos\wordcloud\reimu3.jpg'
img = Image.open(img_path)
img.size
```

```python
import cv2
```

```python
img = cv2.imread(img_path)
type(img)
```

```python
img.shape #numpy三维数组
```

```python
gray_img = numpy.empty(shape = [img.shape[0],img.shape[1]],dtype=int)
```

```python
gray_img.shape
```

```python
for x in range(img.shape[0]):
    for y in range(img.shape[1]):
        R,G,B = tuple(img[x][y].tolist())
        #ndarray对象使用tolist方法转成list，再通过tuple函数转成元组，
        #元组里面元素分别赋值给R G B
        gray_img[x][y] = round(R * 299/1000 + G * 587/1000+ B * 114/1000)
#由于人眼对绿色的敏感最高, 对蓝色敏感最低,
#因此,按上式对RGB三分量进行加权平均能得到较合理的灰度图像，round函数取整
```
