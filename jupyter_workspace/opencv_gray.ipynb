```python
import cv2
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
img = cv2.imread(r'D:\photos\photos\touhou\IMG_009.jpg')
print(type(img))
print(img.shape)
print(img)
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
print(img)
plt.imshow(gray)
#使用cv2.imread()接口读图像，读进来的是BGR格式以及【0～255】。
#所以只要将img转换为RGB格式显示即可,这样图像就不会显示偏绿了,
#但这不适宜用于显示灰度图像 #img = img[:,:,[2,1,0]] GBR转成RGB
```

```python
img = cv2.imread(r'D:\photos\photos\touhou\IMG_009.jpg')
print(img.shape) #(1080, 1920, 3) 三维GBR
#img = img[:,:,[2,1,0]] GBR转成RGB
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
print(gray.shape) #(1080, 1920)一维L灰度图像
plt.imshow(gray,cmap='gray')
#plt显示图像默认三通道，需要在plt.imshow()里添加参数cmap=“gray”
#就可以显示灰度图像了：
```
