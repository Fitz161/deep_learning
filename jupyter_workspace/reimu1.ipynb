```python
from wordcloud import WordCloud, ImageColorGenerator
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import jieba
%matplotlib notebook
```

```python
text_path=r'D:\document\wordcloud_doc\touhou_reimu1.txt'
imgpath=r'D:\photos\wordcloud\reimu4.png'
with open(text_path,errors='ignore',encoding='utf8') as f:
    text=f.read()
text[:100]
```

```python
jieba.add_word("博丽灵梦")
jieba.add_word("八云紫")
jieba.add_word("幻想乡")
words = jieba.lcut(text)
cut=' '.join(words)
print(cut[:100])
```

```python
fontpath=r'D:\uncompress\fangzheng_font\black_font.TTF'
mask=np.array(Image.open(imgpath))
#genclr=ImageColorGenerator(mask)
import random
def random_color(word, font_size, position, orientation, font_path, random_state):
    s = 'hsl(0, %d%%, %d%%)' % (random.randint(60, 80), random.randint(60, 80))
    return s
wc = WordCloud(font_path=fontpath, mode='RGBA',scale=3,mask=mask,color_func=random_color,
               background_color='white').generate(cut)
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
```
